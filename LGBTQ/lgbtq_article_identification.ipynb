```python
import numpy as np
import pandas as pd
import re
import json
from tqdm.auto import tqdm
```

```python
with open('named_entities.json') as f:
    entites = json.load(f)
# entites
```

```python
# df = pd.read_csv('/home/nbook/Projects/Tobacco/TobaccoResearch_Meng/NER/paperAbstractsAll.csv')
df = pd.read_csv('https://docs.google.com/spreadsheets/d/1L30o1enMWRW---ILpZ5_4UzdygQb6tbgFUVobAJ0TI0/edit#gid=0'.replace('/edit#gid=0', '/export?format=csv&gid=0'))
df = df[(df.Vote>0) & (df.Exceptions!=0)]
df.to_csv('lgbtq_articles_clean.csv', index=False)
```

```python
lgbtq = pd.DataFrame(columns=['Conference','Year','Title','PaperType','Matched','Abstract','Highlights','Link','AuthorNames','AuthorAffiliation'])
for idx, row in tqdm(df.iterrows(), total=len(df)):
    abstract = row.Abstract.lower()
    if pd.isna(row.Highlights):
        hightlight = ''
    else:
        hightlight = row.Highlights.lower()
    matched = []
    for entity in entites:
        # re match whole word
        if re.search(r'\b' + entity + r'\b', abstract):
            matched.append(entity)
        elif re.search(r'\b' + entity + r'\b', hightlight):
            matched.append(entity)
    if len(matched) > 0:
        # raise Exception('matched')
        row['Matched'] = matched
        lgbtq.loc[len(lgbtq)] = row
        
```
